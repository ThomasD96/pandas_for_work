```python
import pandas as pd
from pandas.io.parsers import ParserError
```

```python
# create a list to store the dataframes 
dataframe_list = []

# this is a log-files, which stores all infos if the data-import worked or not
file = open('log_file.txt', 'a+')
    

# iterate through the folders 1 to 4 (here: folder_1, folder_2, ... )
# output: 1, 2, 3
for folder in range(1, 4):
    
    # create two boolean variables for both kind of exceptions.
    parse_error = False
    file_not_found = False
    
    ### notice that for folder i, we have file name "data_i"
    ### create the file name
    try :
        file_name = 'data_' + str(folder) + '.csv'   
        # possible to put folders in common sub-folder called 'dataset'
        #data = pd.read_csv('dataset/'+ str(folder) +'/' +file_name, sep = ';')
        data = pd.read_csv(str(folder) +'/' + file_name, sep = ';')
          
    # if the error is ParserError, print file has incorrect format and set parse_error = True
    except ParserError:
        parse_error = True
        data = pd.read_csv('dataset/'+ str(folder) +'/' +file_name,  skiprows=4, sep = ';')
        dataframe_list.append(data)
        print(file_name, 'has incorrect format.')
    
    # if the error is FileNotFoundError, print file is missing and set file_not_found = True
    except FileNotFoundError:
        file_not_found = True
        print(file_name, 'is missing')
    
    # if no exception occurs,  append the dataframe to the list.
    else:
                     
        dataframe_list.append(data)
    # Now, in the end, update the log file.
    finally:
        if parse_error:
            file.writelines(file_name + ' has incorrect format\n')
        
        elif file_not_found:
            file.writelines(file_name + ' is missing.\n')
        
        else:
            file.writelines(file_name + ' is correct.\n')
        
        
file.close()
```
