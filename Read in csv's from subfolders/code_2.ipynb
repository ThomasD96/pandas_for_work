```python
import os
import pandas as pd
from fnmatch import fnmatch
from pandas.io.parsers import ParserError
```

```python
def merge_files_subfolders(axis, pattern):
    
    current_wp = os.getcwd()
    root = current_wp
    pattern = pattern #"*.csv" # like: "*.csv" or "*.xlsx"
    path_links = []
    dataframe_list = []

    for path, subdirs, files in os.walk(root):
        for name in files:
            if fnmatch(name, pattern):
                path_links.append(os.path.join(path, name))
                
                
    if 'csv' in '*.csv':
        for i in path_links:
            data = pd.read_csv(i, sep = ';')
            dataframe_list.append(data)
            
    else:
        for i in path_links:
            data = pd.read_excel(i)
            dataframe_list.append(data)
    
    # axis = 1 ---> column-wise merge
    # axis = 0 ---> row-wise merge
    df = pd.concat(dataframe_list, axis=axis)
    return df
```

```python
data = merge_files_subfolders(axis = 0, pattern = "*.csv")
data
```
