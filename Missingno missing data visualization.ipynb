```python
import missingno.missingno as msno
import pandas as pd
```

```python
df = pd.read_csv('german_credit_data.csv')
```

```python
msno.matrix(df.sample(1000).iloc[:, 0:50], fontsize=16)
```

```python
msno.dendrogram(df)
```

```python
msno.bar(df, color="blue", log=True, figsize=(30,18))
```

```python
msno.heatmap(df, figsize=(20,20))
```
